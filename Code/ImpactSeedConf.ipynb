```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot_nwords(n_topics):
    data = {}
    for conf in (0.5, 0.7, 0.9, 1):
        counts = [x.strip() for x in open("communities/LDA_Eval_Seeded/topics_" + str(n_topics) + "/seeded/500-n_words_per_topic_" + str(conf) + ".txt").readlines()]
        counts = list(map(int, counts))
        data[conf] = counts
    counts = [x.strip() for x in open("communities/LDA_Eval_Seeded/topics_" + str(n_topics) + "/non_seeded/500-n_words_per_topic.txt").readlines()]
    counts = list(map(int, counts))
    data[0] = counts
    for k,v in data.items():
        plt.violinplot(v,  positions=[k], showmeans=True)
    #plt.title("ntopics: " + str(n_topics))
    plt.xlabel("Seed Confidence")
    plt.ylabel("Number of Words per Topic")
    plt.savefig("nwords_sconf.pdf")
```

```python
for i in (100, 200, 300):
    plot_nwords(i)
    plt.show()
```
