```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot_takeovers():
    d = {100: [],
         200: [],
         300: []
        }
    for i in (100, 200, 300):
        counts = [x.strip() for x in open("communities/LDA_Eval_Seeded/topics_" + str(i) + "/non_seeded/500-hashtag_takeover.csv").readlines()]
        counts = list(map(float, counts))
        d[i].append(np.mean(counts))
        
        for conf in (0.5, 0.7, 0.9, 1):
            counts = [x.strip() for x in open("communities/LDA_Eval_Seeded/topics_" + str(i) + "/seeded/500-hashtag_takeover_" + str(conf) + ".csv").readlines()]
            counts = list(map(float, counts))
            d[i].append(np.mean(counts))
    print(d)
    confs = (0, 0.5, 0.7, 0.9, 1)
    ax1 = plt.plot(confs, d[100], ":")
    ax2 = plt.plot(confs, d[200], "--")
    ax3 = plt.plot(confs, d[300], "-.")
    plt.legend((ax1[0], ax2[0], ax3[0]), ("ntopics = 100", "ntopics = 200", "ntopics = 300"))
    plt.xlabel("Seed Confidence")
    plt.ylabel("Mean Hashtag Takeover Rate")
    plt.ylim([-0.05, 0.55])
    plt.savefig("takeover_vs_sconf.pdf")
```

```python
plot_takeovers()
```
